```python
import requests
import pandas as pd
import time

def make_csv(url, name):
    api_url = "https://api.jikan.moe/v4/" + url
    all_data = []
    page = 1

    # 🎯 입력값 받아서 제한 설정 (빈 입력 = 무제한)
    user_input = input(f"💡 '{name}' 데이터를 몇 개까지 수집할까요? (Enter = 전체 수집): ")
    limit = int(user_input) if user_input.strip().isdigit() else None

    if limit:
        print(f"\n📡 {url} 데이터를 최대 {limit}개까지 수집합니다...\n")
    else:
        print(f"\n📡 {url} 데이터를 제한 없이 끝까지 수집합니다...\n")

    while True:
        response = requests.get(f"{api_url}?page={page}")
        data = response.json()

        if "data" not in data or not data["data"]:
            break

        first_item = data["data"][0]

        # ✅ 일반적인 애니 구조 (/anime, /seasons, /top 등)
        if "title" in first_item:
            for anime in data["data"]:
                all_data.append({
                    "Title": anime.get("title"),
                    "Type": anime.get("type"),
                    "Episodes": anime.get("episodes"),
                    "Year": anime.get("year", "Unknown"),
                    "Score": anime.get("score"),
                    "Genres": ", ".join([g["name"] for g in anime.get("genres", [])]),
                    "URL": anime.get("url")
                })
                if limit and len(all_data) >= limit:
                    break

        # ✅ 추천 구조 (/recommendations/anime)
        elif "entry" in first_item:
            for rec in data["data"]:
                entries = rec.get("entry", [])
                if len(entries) >= 2:
                    all_data.append({
                        "Anime_1": entries[0].get("title"),
                        "Anime_2": entries[1].get("title"),
                        "Votes": rec.get("votes"),
                        "Anime_1_URL": entries[0].get("url"),
                        "Anime_2_URL": entries[1].get("url")
                    })
                    if limit and len(all_data) >= limit:
                        break

        else:
            print("⚠️ 인식되지 않은 데이터 구조:", first_item.keys())
            break

        print(f"📄 {page}페이지 수집 완료 ({len(all_data)}개 누적)")

        # ✅ 조건: 제한 or 다음 페이지 없음
        if (limit and len(all_data) >= limit) or not data["pagination"]["has_next_page"]:
            break

        page += 1
        time.sleep(1)

    # ✅ CSV 저장
    df = pd.DataFrame(all_data)
    df.to_csv(f"{name}_anime_full.csv", index=False, encoding="utf-8-sig")

    print(f"\n✅ '{name}_anime_full.csv' 저장 완료! 총 {len(df)}개의 데이터 수집 완료.\n")
```

```python
make_csv("recommendations/anime", "recommend")
```

```python
make_csv("top/anime", "top")
```

```python
make_csv("seasons/now", "now")
```

```python
make_csv("seasons/upcoming", "upcoming")
```

```python
url = "https://api.jikan.moe/v4/genres/anime"
response = requests.get(url)
data = response.json()

genres = data.get("data", [])

genre_list = []
for genre in genres:
    genre_list.append({
        "mal_id": genre.get("mal_id"),
        "name": genre.get("name"),
        "url": genre.get("url"),
        "count": genre.get("count")  # 작품 수 (있다면)
    })

df = pd.DataFrame(genre_list)
df.to_csv("jikan_genres.csv", index=False, encoding="utf-8-sig")

print(f"✅ 총 {len(df)}개의 장르가 저장되었습니다: jikan_genres.csv")
```
